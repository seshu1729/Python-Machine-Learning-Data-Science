```python
import pandas as pd

# Create a sample DataFrame
data = {'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue'],
        'Size': ['Small', 'Medium', 'Large', 'Small', 'Medium']}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("\n")

# Example with drop_first=False (default behavior)
# This creates a dummy variable for each unique category
df_dummies_false = pd.get_dummies(df, drop_first=False, dtype=int)
print("DataFrame with drop_first=False:")
print(df_dummies_false)
print("\n")

# Example with drop_first=True
# This drops the first category alphabetically for each column
df_dummies_true = pd.get_dummies(df, drop_first=True, dtype=int)
print("DataFrame with drop_first=True:")
print(df_dummies_true)
```
